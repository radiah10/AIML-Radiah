```python
# how to use break, continue, input validation, and exception handling
while True:
    num = input("enter a number") # string

    # check for exit
    if (num == "quit") or (num == 'q'):
        print("Program Stopped!")
        break

    # validate if input is an integer
    # consider all wrong things users can send
    # alphabets or words or symbols or numbers
    if not num.isdigit():
        print("Not an INTEGER. Try again.")
        continue

    # Exception Handling - if we did not have input validation
    try:
        print("Square is: ", int(num) **2)
        print("Cube is: ", int(num) **3)
        print("Square Root is: ", int(num) **0.5)
    except ValueError:
        print("That is not a valid Input. Enter a number.")
```

```python
# take an input from the user
# let's say we have to do a google search
# or we have to prompt an LLM - ChatGPT, or Gemini etc

prompt = input("Enter your query:")
print(prompt)

# string operations help us clean the users inputs
```

```python
print(prompt)

print(prompt.lstrip()) # removes the leading spaces - LEFT
print(prompt.rstrip()) # RIGHT - removes trailing spaces

print(prompt.strip()) # removes useless spaces on left or right

print(prompt.upper())
print(prompt.lower())

# search for something
prompt.find("awesome") # -1 if not found
prompt.find('who') # returns the index of the word

prompt.count("w")
prompt.count("who")

prompt.replace('is', 'are') # find a word and replace it

# we use replace to clean text data
# stop words - remove stopwords

prompt.split() # returns each word broken from whitespaces
# it returns a list

prompt.split(".") # if you have a paragraph and you want to break it into sentences

# check whether something in in a string or not
# membership testing

"what" in prompt  # in --- is used to test membership
"who" in prompt 

prompt.title() # the first letter is capitalized
```
