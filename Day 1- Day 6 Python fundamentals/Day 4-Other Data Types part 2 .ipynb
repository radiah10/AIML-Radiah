```python
set_A = {1,2,3,4,4,3,2}
print(set_A)


set_B = {3,4,8,5,9,6}
print(set_B)
```

```python
# Set Operations
set_A | set_B # Union - is combination of all elements uniquely

set_A & set_B # Intersection - the common elements in the sets
```
