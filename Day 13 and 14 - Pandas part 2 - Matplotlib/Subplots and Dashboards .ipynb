```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataset for comparison
data = pd.DataFrame({
    'Age': [23, 45, 34, 42, 25, 36, 52, 39, 40, 30],
    'Salary': [40000, 80000, 60000, 75000, 42000, 62000, 95000, 71000, 85000, 53000]
})

# Scatter Plot of Age vs Salary
sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple')
plt.title('Age vs Salary (Scatter Plot)')
plt.show()

# Histogram of Age
sns.histplot(data['Age'], bins=5, kde=True, color='green')
plt.title('Histogram of Age')
plt.show()

# Box Plot of Salary
sns.boxplot(data['Salary'], color='orange')
plt.title('Box Plot of Salary')
plt.show()
```

```python
# Create a figure with 1 row and 3 columns
plt.figure(figsize=(15, 5)) # This creates a figure with a specific size, here 15 inches wide and 5 inches tall, to accommodate three plots side by side.

# Scatter Plot (1st subplot)
plt.subplot(1, 3, 1)  # (rows, columns, position)
sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple')
plt.title('Age vs Salary (Scatter Plot)')
plt.xlabel('Age')
plt.ylabel('Salary')

# Histogram (2nd subplot)
plt.subplot(1, 3, 2)  # (rows, columns, position)
sns.histplot(data['Age'], bins=5, kde=True, color='green')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Box Plot (3rd subplot)
plt.subplot(1, 3, 3)  # (rows, columns, position)
sns.boxplot(data['Salary'], color='orange')
plt.title('Box Plot of Salary')
plt.xlabel('Salary')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
plt.figure(figsize= (12, 4)) # width, height

plt.subplot(1,3, 1) # nrows, ncols, index
sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple')


plt.subplot(1,3, 2)
sns.histplot(data['Age'], bins=5, kde=True, color='green')

plt.subplot(1,3, 3)
sns.boxplot(data['Salary'], color='orange')
```

```python
fig, ax = plt.subplots(1,3, figsize = (12, 5)) # creates a canvas

sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple', ax = ax[0])
ax[0].set_title('Age vs Salary (Scatter Plot)')
ax[0].set_xlabel('Age')
ax[0].set_ylabel('Salary')

sns.histplot(data['Age'], bins=5, kde=True, color='green', ax = ax[1])
ax[1].set_title('Age Distribution (Histogram)')
ax[1].set_xlabel('Age')
ax[1].set_ylabel('Frequency')

sns.boxplot(data['Salary'], color='orange', ax = ax[2])
ax[2].set_title('Salary Distribution (Box Plot)')
ax[2].set_xlabel('Salary')
ax[2].set_ylabel('Salary')


plt.suptitle("Salary and Age Dashboard", 
             fontsize = 24,
             fontweight = 'bold',
             y = 1.03)

plt.tight_layout() # prevents overlaps
```

```python
fig, ax = plt.subplots(2,3, figsize = (12, 7)) # creates a canvas

sns.set_theme(style= 'darkgrid', 
              palette= 'pastel', #'pastel',
              font= 'sans-serif',
              font_scale= 1.1,)

sns.scatterplot(x=data['Age'], y=data['Salary'], ax = ax[0,0])
ax[0,0].set_title('Age vs Salary (Scatter Plot)')
ax[0,0].set_xlabel('Age')
ax[0,0].set_ylabel('Salary')

sns.histplot(data['Age'], bins=5, kde=True, ax = ax[0,1])
ax[0,1].set_title('Age Distribution (Histogram)')
ax[0,1].set_xlabel('Age')
ax[0,1].set_ylabel('Frequency')

sns.boxplot(data['Salary'], ax = ax[0, 2])
ax[0,2].set_title('Salary Distribution (Box Plot)')
ax[0,2].set_xlabel('Salary')
ax[0,2].set_ylabel('Salary')


plt.suptitle("Salary and Age Dashboard", 
             fontsize = 24,
             fontweight = 'bold',
             y = 1.03)

plt.tight_layout() # prevents overlaps
plt.show()
```
