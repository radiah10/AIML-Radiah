```python
import numpy as np

# A NumPy array of 64-bit integers (8 bytes each)
np_arr = np.array([100, 101, 102], dtype=np.int64)

print("--- NumPy Array Memory ---")
print(f"Base memory address: {np_arr.ctypes.data}")
print(f"Step size (strides): {np_arr.strides}") 

# Let's calculate the address of each element manually
base = np_arr.ctypes.data

for i in range(len(np_arr)):
    print(f"Element {i} address: {base + (i * 8)}")
```
