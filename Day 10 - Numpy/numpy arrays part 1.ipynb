```python
arr_table = np.array([[1, 2, 3, 4],
                    [3, 4, 5, 6],
                    [3, 4, 5, 6],
                    [3, 4, 5, 6],
                    [3, 4, 5, 6],
                    [3, 4, 5, 6],
                    [3, 4, 5, 6],
                    [3, 4, 5, 6],
                    [3, 4, 5, 6],
                    [3, 4, 5, 6],]) # 2D array


arr_table.ndim # 2

arr_table.shape # (10, 4) ---> 10 rows, 4 columns
```

```python
# please take all elements and multiply them by 100
arr_table * 10

# for each column different values to operate
# first column every element must be taken in 10s, 2nd --> 100s, 3rd 1000s, 4th ---> 10000

# (10, 4) --->  

arr_table * [10,100,1000,10000] # (10, 4) --- (4,) =====  10 x 4    --- 1 x 4

# mis match of shapes
# arr_table * [10,100,1000,10000,10,100,1000,10,100,1000] # (10,4) (10,)   ===== 10 x 4 ---- 1 x 10 (not possible)

arr_table * [[10],[100],[1000],[10000],[10],[100],[1000],[10],[100],[1000]]  #  (10, 4) --- (10, 1)
```

```python
# an array of 12 elements
arr = np.arange(12)
print("Original array", arr)


# reshape into 3,4
arr_34 = arr.reshape(3,4) # 4,3 --- 2,6 ---6,2 -- 
print("Reshaped array\n", arr_34)

# flatten - converts to 1D
arr_flattened = arr_34.flatten() # ravel() - does the same thing, take an array and flatten it
print("Flattened\n", arr_flattened)

# transpose --- swap the rows and columns
# fliping a credit card while keeping two diagonal corners in your fingers
print("transposed array\n", arr_34.T)
```

```python
arr_34

arr_34.argmax() # index of highest element in the array

arr_34.argmin() # index of lowest element in the array

arr_34.ravel() # flattens array
```

```python
arr_a = np.array([[ 10,  2,   3],
                  [ 4,   5,   6],
                  [ 7,   8,  90]])


print("Row 0 ===>", arr_a[0])
print("Row 0, Column 0 ===>", arr_a[0,0])
print("Row 1, Column 2 ===>", arr_a[1,2])

arr_a[1,2] = 100
print("after modification = Row 1, Column 2 ===>", arr_a[1,2])


arr_a[0][0] # we use this method in lists as well

# arrays have something extra too
# INDEXING - targets (row,col)
arr_a[0,0]


# subset of the array
# sub-matrix
# CROP opeartion of images
arr_a[1:3 , 0:2] # take row 1, 2 , take col 0, 1
```
